## Ejercicio 1 · Máximo Común Divisor (algoritmo de Euclides)

Implementa `mcd(a, b)` que devuelva el Máximo Común Divisor de dos enteros positivos usando el **algoritmo de Euclides**: mientras `b != 0`, reemplaza `(a, b)` por `(b, a % b)`; cuando `b == 0`, devuelve `a`.

**Ejemplos:** `mcd(48, 18)` → `6`; `mcd(100, 75)` → `25`; `mcd(13, 7)` → `1`.

```python
def mcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
```

```python
# === Tests visibles · Ejercicio 1 ===
assert mcd(48, 18) == 6
assert mcd(100, 75) == 25
assert mcd(13, 7) == 1
print("✅ Ejercicio 1: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 1 ===
assert mcd(1, 1) == 1
assert mcd(252, 105) == 21
assert mcd(17, 17) == 17
assert mcd(36, 24) == 12
print("✅ Ejercicio 1: tests adicionales superados.")
```

## Ejercicio 2 · Bubble sort (ordenamiento propio)

Implementa `bubble_sort(lista)` que ordene una lista de números de **menor a mayor** usando el algoritmo de burbuja: recorre la lista repetidamente e intercambia elementos adyacentes que estén en el orden incorrecto. Devuelve una **nueva lista** (no modifica la original).

**Ejemplo:** `bubble_sort([5, 2, 4, 1, 3])` → `[1, 2, 3, 4, 5]`.

Pista: una pasada completa 'burbujea' el mayor al final. Repite `n-1` pasadas.

```python
def bubble_sort(lista):
    arr = list(lista)   # copia para no mutar la original
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

```python
# === Tests visibles · Ejercicio 2 ===
assert bubble_sort([5, 2, 4, 1, 3]) == [1, 2, 3, 4, 5]
assert bubble_sort([1]) == [1]
assert bubble_sort([]) == []
print("✅ Ejercicio 2: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 2 ===
assert bubble_sort([3, 3, 1]) == [1, 3, 3]
assert bubble_sort([-2, 0, -5, 3]) == [-5, -2, 0, 3]
_orig = [4, 2, 1]; bubble_sort(_orig); assert _orig == [4, 2, 1]
print("✅ Ejercicio 2: tests adicionales superados.")
```

## Ejercicio 3 · Contar ocurrencias de cada letra

Implementa `contar_letras(texto)` que reciba una cadena (solo letras minúsculas y espacios) y devuelva una **lista de tuplas** `(letra, conteo)` ordenada **alfabéticamente** por letra. Ignora los espacios.

**Ejemplo:** `contar_letras("abba")` → `[('a', 2), ('b', 2)]`.

Pista: construye dos listas paralelas (letras únicas + conteos), luego ordena.

```python
def contar_letras(texto):
    letras = []
    conteos = []
    for c in texto:
        if c == ' ':
            continue
        if c in letras:
            conteos[letras.index(c)] += 1
        else:
            letras.append(c)
            conteos.append(1)
    pares = list(zip(letras, conteos))
    return sorted(pares, key=lambda p: p[0])
```

```python
# === Tests visibles · Ejercicio 3 ===
assert contar_letras('abba') == [('a', 2), ('b', 2)]
assert contar_letras('aaa') == [('a', 3)]
print("✅ Ejercicio 3: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 3 ===
assert contar_letras('') == []
assert contar_letras('cba') == [('a', 1), ('b', 1), ('c', 1)]
assert contar_letras('hola mundo') == [('a', 1), ('d', 1), ('h', 1), ('l', 1), ('m', 1), ('n', 1), ('o', 2), ('u', 1)]
print("✅ Ejercicio 3: tests adicionales superados.")
```

## Ejercicio 4 · Números primos hasta n (criba manual)

Implementa `primos_hasta(n)` que devuelva una **lista** con todos los números primos desde 2 hasta `n` inclusive. Un número es primo si solo es divisible por 1 y por sí mismo.

**Ejemplo:** `primos_hasta(20)` → `[2, 3, 5, 7, 11, 13, 17, 19]`.

Pista: escribe una función auxiliar `es_primo(k)` y usa un bucle externo.

```python
def primos_hasta(n):
    def es_primo(k):
        if k < 2:
            return False
        d = 2
        while d * d <= k:
            if k % d == 0:
                return False
            d += 1
        return True
    return [k for k in range(2, n + 1) if es_primo(k)]
```

```python
# === Tests visibles · Ejercicio 4 ===
assert primos_hasta(20) == [2, 3, 5, 7, 11, 13, 17, 19]
assert primos_hasta(2) == [2]
assert primos_hasta(1) == []
print("✅ Ejercicio 4: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 4 ===
assert primos_hasta(10) == [2, 3, 5, 7]
assert 97 in primos_hasta(100)
assert 100 not in primos_hasta(100)
print("✅ Ejercicio 4: tests adicionales superados.")
```

## Ejercicio 5 · Aplicar descuentos con map y lambda

Implementa `aplicar_descuentos(precios, tasa)` que reciba una lista de precios y una tasa de descuento (ej: `0.10` = 10%) y devuelva una **nueva lista** con los precios tras el descuento, usando `map` y `lambda`.

Los precios resultado deben ser enteros (usa `int()` en la lambda).

**Ejemplo:** `aplicar_descuentos([100000, 200000], 0.10)` → `[90000, 180000]`.

```python
def aplicar_descuentos(precios, tasa):
    return list(map(lambda p: int(p * (1 - tasa)), precios))
```

```python
# === Tests visibles · Ejercicio 5 ===
assert aplicar_descuentos([100000, 200000], 0.10) == [90000, 180000]
assert aplicar_descuentos([50000], 0.50) == [25000]
print("✅ Ejercicio 5: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 5 ===
assert aplicar_descuentos([], 0.20) == []
assert aplicar_descuentos([10000, 20000, 30000], 0.0) == [10000, 20000, 30000]
assert aplicar_descuentos([99999], 0.19) == [80999]
print("✅ Ejercicio 5: tests adicionales superados.")
```

## Ejercicio 6 · Potencias de 2 hasta 2^n

Implementa `potencias_de_2(n)` que devuelva una **lista** con las potencias de 2 desde 2^0 hasta 2^n inclusive.

**Ejemplo:** `potencias_de_2(5)` → `[1, 2, 4, 8, 16, 32]`.

La lista tiene `n + 1` elementos.

```python
def potencias_de_2(n):
    resultado = []
    valor = 1
    for _ in range(n + 1):
        resultado.append(valor)
        valor *= 2
    return resultado
```

```python
# === Tests visibles · Ejercicio 6 ===
assert potencias_de_2(5) == [1, 2, 4, 8, 16, 32]
assert potencias_de_2(0) == [1]
print("✅ Ejercicio 6: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 6 ===
assert potencias_de_2(1) == [1, 2]
assert potencias_de_2(10)[-1] == 1024
assert len(potencias_de_2(8)) == 9
print("✅ Ejercicio 6: tests adicionales superados.")
```

## Ejercicio 7 · Acumular comisión por tramos de ventas

Implementa `comision_total(ventas)` que reciba una lista de montos de venta y devuelva la **comisión total acumulada**. La tasa de comisión depende de cada venta individual:
- venta < 500.000 → 5%
- 500.000 <= venta <= 2.000.000 → 8%
- venta > 2.000.000 → 12%

**Ejemplo:** `comision_total([200000, 800000, 3000000])` → `10000 + 64000 + 360000 = 434000`.

```python
def comision_total(ventas):
    def tasa(v):
        if v < 500000:
            return 0.05
        elif v <= 2000000:
            return 0.08
        else:
            return 0.12
    total = 0
    for v in ventas:
        total += v * tasa(v)
    return total
```

```python
# === Tests visibles · Ejercicio 7 ===
assert comision_total([200000, 800000, 3000000]) == 434000.0
assert comision_total([]) == 0
print("✅ Ejercicio 7: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 7 ===
assert comision_total([100000]) == 5000.0
assert comision_total([500000]) == 40000.0
assert comision_total([2000001]) == 240000.12
assert abs(comision_total([499999, 500000, 2000000, 2000001]) - (499999*0.05 + 500000*0.08 + 2000000*0.08 + 2000001*0.12)) < 0.01
print("✅ Ejercicio 7: tests adicionales superados.")
```

## Ejercicio 8 · Validar condición sobre todos/alguno de los elementos

Implementa dos funciones:
- `todos_cumplen(lista, condicion)`: devuelve `True` si **todos** los elementos de la lista cumplen `condicion`. Lista vacía → `True`.
- `alguno_cumple(lista, condicion)`: devuelve `True` si **al menos un** elemento cumple `condicion`. Lista vacía → `False`.

Implementa internamente con un **bucle** (no uses `all()` ni `any()` directamente en la función principal).

**Ejemplos:**
`todos_cumplen([2, 4, 6], lambda x: x % 2 == 0)` → `True`
`alguno_cumple([1, 3, 4], lambda x: x % 2 == 0)` → `True`

```python
def todos_cumplen(lista, condicion):
    for x in lista:
        if not condicion(x):
            return False
    return True

def alguno_cumple(lista, condicion):
    for x in lista:
        if condicion(x):
            return True
    return False
```

```python
# === Tests visibles · Ejercicio 8 ===
assert todos_cumplen([2, 4, 6], lambda x: x % 2 == 0) is True
assert todos_cumplen([2, 3, 6], lambda x: x % 2 == 0) is False
assert alguno_cumple([1, 3, 4], lambda x: x % 2 == 0) is True
assert alguno_cumple([1, 3, 5], lambda x: x % 2 == 0) is False
print("✅ Ejercicio 8: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 8 ===
assert todos_cumplen([], lambda x: False) is True
assert alguno_cumple([], lambda x: True) is False
assert todos_cumplen([10, 20, 30], lambda x: x > 5) is True
assert alguno_cumple([-1, -2, 3], lambda x: x > 0) is True
print("✅ Ejercicio 8: tests adicionales superados.")
```
