## Ejercicio 1 · Días entre dos años

Implementa `dias_entre_anios(anio_inicio, anio_fin)` que devuelva el número total de días entre el 1 de enero de `anio_inicio` y el 1 de enero de `anio_fin`.

**Sin librerías de fecha.** Usa aritmética: recorre cada año y suma 366 si es bisiesto, 365 si no.

Un año es bisiesto si: `(anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0)`.

**Ejemplo:** `dias_entre_anios(2020, 2023)` → `1096`
(2020 tiene 366 días + 2021 tiene 365 + 2022 tiene 365 = 1096).

```python
def dias_entre_anios(anio_inicio, anio_fin):
    total = 0
    for a in range(anio_inicio, anio_fin):
        bisiesto = (a % 4 == 0 and a % 100 != 0) or (a % 400 == 0)
        total += 366 if bisiesto else 365
    return total
```

```python
# === Tests visibles · Ejercicio 1 ===
assert dias_entre_anios(2020, 2023) == 1096
assert dias_entre_anios(2023, 2024) == 365
print("✅ Ejercicio 1: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 1 ===
assert dias_entre_anios(2000, 2001) == 366
assert dias_entre_anios(1900, 1901) == 365
assert dias_entre_anios(2024, 2024) == 0
print("✅ Ejercicio 1: tests adicionales superados.")
```

## Ejercicio 2 · Tipo de triángulo

Implementa `tipo_triangulo(a, b, c)` que clasifique el triángulo según sus lados:
- `'equilatero'` si los tres lados son iguales.
- `'isosceles'` si exactamente dos lados son iguales.
- `'escaleno'` si ningún lado es igual a otro.
- `'invalido'` si los lados no forman un triángulo (la suma de dos lados cualquiera debe ser mayor al tercero).

**Ejemplos:** `tipo_triangulo(5, 5, 5)` → `'equilatero'`; `tipo_triangulo(3, 3, 5)` → `'isosceles'`; `tipo_triangulo(3, 4, 5)` → `'escaleno'`; `tipo_triangulo(1, 2, 10)` → `'invalido'`.

```python
def tipo_triangulo(a, b, c):
    if a + b <= c or a + c <= b or b + c <= a:
        return 'invalido'
    if a == b == c:
        return 'equilatero'
    elif a == b or b == c or a == c:
        return 'isosceles'
    else:
        return 'escaleno'
```

```python
# === Tests visibles · Ejercicio 2 ===
assert tipo_triangulo(5, 5, 5) == 'equilatero'
assert tipo_triangulo(3, 3, 5) == 'isosceles'
assert tipo_triangulo(3, 4, 5) == 'escaleno'
assert tipo_triangulo(1, 2, 10) == 'invalido'
print("✅ Ejercicio 2: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 2 ===
assert tipo_triangulo(0, 0, 0) == 'invalido'
assert tipo_triangulo(7, 7, 9) == 'isosceles'
assert tipo_triangulo(6, 8, 10) == 'escaleno'
assert tipo_triangulo(1, 1, 2) == 'invalido'
print("✅ Ejercicio 2: tests adicionales superados.")
```

## Ejercicio 3 · Índice de rentabilidad

Implementa `rentabilidad(precio_compra, precio_venta)` que devuelva:
- El **porcentaje de ganancia o pérdida** sobre el precio de compra,
  redondeado a 4 decimales.
- Formula: `(precio_venta - precio_compra) / precio_compra`

- Si `precio_compra <= 0`, devuelve `None`.

**Ejemplos:** `rentabilidad(1000000, 1200000)` → `0.2` (ganancia 20%);
`rentabilidad(1000000, 800000)` → `-0.2` (pérdida 20%).

```python
def rentabilidad(precio_compra, precio_venta):
    if precio_compra <= 0:
        return None
    return round((precio_venta - precio_compra) / precio_compra, 4)
```

```python
# === Tests visibles · Ejercicio 3 ===
assert rentabilidad(1000000, 1200000) == 0.2
assert rentabilidad(1000000, 800000) == -0.2
print("✅ Ejercicio 3: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 3 ===
assert rentabilidad(500000, 500000) == 0.0
assert rentabilidad(0, 100000) is None
assert rentabilidad(200000, 250000) == 0.25
print("✅ Ejercicio 3: tests adicionales superados.")
```

## Ejercicio 4 · Clasificar nota universitaria

Implementa `clasificar_nota(nota)` que devuelva la clasificación de una nota:
- `'reprobado'` si `nota < 3.0`
- `'aprobado'` si `3.0 <= nota < 4.0`
- `'bueno'` si `4.0 <= nota < 4.5`
- `'excelente'` si `nota >= 4.5`

**Nota:** la escala es de 0 a 5. Si la nota está fuera de `[0, 5]`, devuelve `'fuera de rango'`.

**Ejemplos:** `clasificar_nota(2.5)` → `'reprobado'`; `clasificar_nota(3.8)` → `'aprobado'`; `clasificar_nota(4.7)` → `'excelente'`.

```python
def clasificar_nota(nota):
    if nota < 0 or nota > 5:
        return 'fuera de rango'
    if nota < 3.0:
        return 'reprobado'
    elif nota < 4.0:
        return 'aprobado'
    elif nota < 4.5:
        return 'bueno'
    else:
        return 'excelente'
```

```python
# === Tests visibles · Ejercicio 4 ===
assert clasificar_nota(2.5) == 'reprobado'
assert clasificar_nota(3.8) == 'aprobado'
assert clasificar_nota(4.7) == 'excelente'
print("✅ Ejercicio 4: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 4 ===
assert clasificar_nota(0.0) == 'reprobado'
assert clasificar_nota(3.0) == 'aprobado'
assert clasificar_nota(4.0) == 'bueno'
assert clasificar_nota(4.5) == 'excelente'
assert clasificar_nota(5.0) == 'excelente'
assert clasificar_nota(-1) == 'fuera de rango'
assert clasificar_nota(6) == 'fuera de rango'
print("✅ Ejercicio 4: tests adicionales superados.")
```

## Ejercicio 5 · Interés simple

Implementa `interes_simple(capital, tasa, tiempo)` que devuelva una tupla `(interes, monto_total)` donde:
- `interes = capital * tasa * tiempo`
- `monto_total = capital + interes`

Los valores se redondean a 2 decimales.

Si `capital <= 0`, `tasa < 0` o `tiempo < 0`, devuelve `(None, None)`.

**Ejemplo:** `interes_simple(1000000, 0.05, 3)` → `(150000.0, 1150000.0)`
(capital de 1M, tasa 5% anual, 3 años).

```python
def interes_simple(capital, tasa, tiempo):
    if capital <= 0 or tasa < 0 or tiempo < 0:
        return (None, None)
    interes = round(capital * tasa * tiempo, 2)
    return (interes, round(capital + interes, 2))
```

```python
# === Tests visibles · Ejercicio 5 ===
assert interes_simple(1000000, 0.05, 3) == (150000.0, 1150000.0)
assert interes_simple(500000, 0.10, 1) == (50000.0, 550000.0)
print("✅ Ejercicio 5: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 5 ===
assert interes_simple(0, 0.05, 3) == (None, None)
assert interes_simple(1000000, 0, 3) == (0.0, 1000000.0)
assert interes_simple(1000000, 0.05, 0) == (0.0, 1000000.0)
assert interes_simple(-1000, 0.05, 3) == (None, None)
print("✅ Ejercicio 5: tests adicionales superados.")
```

## Ejercicio 7 · Precio con descuento e IVA encadenados

Implementa `precio_final(precio_original, descuento, tasa_iva)` que:
1. Aplica el descuento al precio original: `precio_desc = precio_original * (1 - descuento)`
2. Aplica el IVA sobre el precio con descuento: `final = precio_desc * (1 + tasa_iva)`
3. Devuelve una tupla `(precio_desc, final)` redondeada a 2 decimales.

Si `precio_original <= 0` o `descuento < 0` o `descuento > 1` o `tasa_iva < 0`, devuelve `(None, None)`.

**Ejemplo:** `precio_final(100000, 0.10, 0.19)` → `(90000.0, 107100.0)`.

```python
def precio_final(precio_original, descuento, tasa_iva):
    if precio_original <= 0 or descuento < 0 or descuento > 1 or tasa_iva < 0:
        return (None, None)
    precio_desc = round(precio_original * (1 - descuento), 2)
    final = round(precio_desc * (1 + tasa_iva), 2)
    return (precio_desc, final)
```

```python
# === Tests visibles · Ejercicio 7 ===
assert precio_final(100000, 0.10, 0.19) == (90000.0, 107100.0)
assert precio_final(50000, 0, 0.19) == (50000.0, 59500.0)
print("✅ Ejercicio 7: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 7 ===
assert precio_final(0, 0.10, 0.19) == (None, None)
assert precio_final(100000, -0.1, 0.19) == (None, None)
assert precio_final(100000, 1.5, 0.19) == (None, None)
assert precio_final(200000, 0.15, 0.08) == (170000.0, 183600.0)
print("✅ Ejercicio 7: tests adicionales superados.")
```

## Ejercicio 8 · Convertir segundos a horas, minutos y segundos

Implementa `segundos_a_hms(total_segundos)` que convierta un número de segundos a una tupla `(horas, minutos, segundos)`.

Usa solo división entera `//` y módulo `%`. Si `total_segundos < 0`, devuelve `(None, None, None)`.

**Ejemplos:**
- `segundos_a_hms(3661)` → `(1, 1, 1)` (1h 1m 1s)
- `segundos_a_hms(90)` → `(0, 1, 30)` (0h 1m 30s)
- `segundos_a_hms(0)` → `(0, 0, 0)`

```python
def segundos_a_hms(total_segundos):
    if total_segundos < 0:
        return (None, None, None)
    horas   = total_segundos // 3600
    resto   = total_segundos % 3600
    minutos = resto // 60
    segs    = resto % 60
    return (horas, minutos, segs)
```

```python
# === Tests visibles · Ejercicio 8 ===
assert segundos_a_hms(3661) == (1, 1, 1)
assert segundos_a_hms(90)   == (0, 1, 30)
assert segundos_a_hms(0)    == (0, 0, 0)
print("✅ Ejercicio 8: tests visibles superados.")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 8 ===
assert segundos_a_hms(3600) == (1, 0, 0)
assert segundos_a_hms(86399) == (23, 59, 59)
assert segundos_a_hms(-1) == (None, None, None)
assert segundos_a_hms(9753) == (2, 42, 33)
print("✅ Ejercicio 8: tests adicionales superados.")
```
