````python
def revisar(nombre: str, condicion: bool, pista: str = "") -> bool:
    """Imprime ✅ o ❌ según `condicion`. Devuelve el booleano para encadenar.

    Parameters
    ----------
    nombre : str
        Descripción legible de lo que se está probando.
    condicion : bool
        Resultado de la comprobación (normalmente una comparación ``==``).
    pista : str, opcional
        Mensaje de ayuda que se muestra solo cuando la prueba falla.
    """
    if condicion:
        print(f"✅ {nombre}")
        return True
    msg = f"❌ {nombre}"
    if pista:
        msg += f"  →  pista: {pista}"
    print(msg)
    return False
````

## Ejercicio 1 · Área de un rectángulo (Entrada → Proceso → Salida)

Escribe `area_rectangulo(base, altura)` que devuelva el área.

**Ejemplo:** `area_rectangulo(3, 4)` → `12`.

Es el patrón EPS más simple: entran dos números, sale uno.

Pseudocodigo:
Leer base y altura
SI base es menor que 0 O altura es menor que 0 ENTONCES
  retornar: "valores deben ser positivos"
SI NO
  area = base * altura
  retornar: area
  

````python
def area_rectangulo(base, altura):
    if base < 0 or altura < 0:
        return "valores deben ser positivos"

    area = base * altura

    return area
````

````python
revisar("3 x 4 = 12", area_rectangulo(3, 4) == 12)
revisar("10 x 2 = 20", area_rectangulo(10, 2) == 20)
revisar("0 de base = 0", area_rectangulo(0, 5) == 0)
````

secuencia.

## Ejercicio 2 · El mayor de tres números (decisión)

Escribe `mayor_de_tres(a, b, c)` que devuelva el más grande de los tres.

**Ejemplo:** `mayor_de_tres(4, 9, 2)` → `9`.

Pista: usa comparaciones. (Puedes usar `max`, pero intenta primero con `if`.)

Recibe los tres numeros a b c
Si a es mayor o igual que b y c es mayor o igual que c entonces retornar a
Si no, si b es mayor o igual que a y b es mayor o igual que c retorne b Si no: retorne c

````python
def mayor_de_tres(a, b, c):
    if a >= b and a >= c:
      return a
    elif b >= a and b >= c:
      return b
    else:
          return c
````

````python
revisar("4,9,2 -> 9", mayor_de_tres(4, 9, 2) == 9)
revisar("el primero", mayor_de_tres(10, 1, 5) == 10)
revisar("negativos", mayor_de_tres(-3, -1, -7) == -1)
````

## Ejercicio 3 · Máximo de una lista (patrón campeón)

Escribe `maximo(lista)` que devuelva el mayor valor de una lista **no vacía**,
sin usar `max`.

**Ejemplo:** `maximo([3, 9, 2, 9, 1])` → `9`.

Guardar el primer elemento de la lista como el mayor
Recorrer cada elemento de la lista  Si el dato actual es mayor que el valor mayor entonces Actualizar el valor mayor con el elemento actual retornar el valor mayor
    

````python
def maximo(lista):
    ganador = lista[0]    #Posicion 0 de la lista Python incia desde cero
    for x in lista:
        if x > ganador:
            ganador = x
    return ganador
````

````python
revisar("ejemplo", maximo([3, 9, 2, 9, 1]) == 9)
revisar("un solo elemento", maximo([7]) == 7)
revisar("todos negativos", maximo([-5, -2, -9]) == -2)
````


## Ejercicio 4 · Contar números pares (contador)

Escribe `contar_pares(lista)` que cuente cuántos números pares hay.

**Ejemplo:** `contar_pares([1, 2, 3, 4, 6])` → `3`.

PSEUDOCODIGO: Iniciar contador en 0, recorrer cada elemento de la lista, si el dato actual dividido 2 es igual a cero sumar uno al contador, retornar el contador.

````python
def contar_pares(lista):
    contar = 0
    for x in lista:
        if x % 2 == 0:
            contar = contar + 1
    return contar
````

````python
revisar("ejemplo", contar_pares([1, 2, 3, 4, 6]) == 3)
revisar("ninguno", contar_pares([1, 3, 5]) == 0)
revisar("lista vacía", contar_pares([]) == 0)
````

## Ejercicio 5 · Promedio seguro (acumulador + caso borde)

Escribe `promedio(lista)` que devuelva el promedio. Si la lista está **vacía**,
devuelve `None` (no se puede dividir entre cero).

**Ejemplo:** `promedio([4, 6, 8])` → `6.0`; `promedio([])` → `None`.

PSEUDOCODIGO: SI la lista esta vadio retornar None SI NO: sumar todos los valores de lista contar la cantidad de elementos de la lista, calcular promedio  igual suma / cantidad de elementos retornar  promedio

````python
def promedio(lista):
    if len(lista) == 0:    #Len es una funcion de python que nos dice cuantos elementos tiene algo
        return None
    suma = 0
    for x in lista:
        suma = suma + x
    return suma / len(lista)
````

````python
revisar("ejemplo", promedio([4, 6, 8]) == 6.0)
revisar("un valor", promedio([10]) == 10.0)
revisar("vacía -> None", promedio([]) is None)
````

<details><summary>💡 Ver solución</summary>

```python
def promedio(lista):
    if len(lista) == 0:      # validar ANTES de dividir
        return None
    total = 0
    for x in lista:
        total += x           # acumulador
    return total / len(lista)
```
Primero validamos el caso borde (lista vacía), luego acumulamos y dividimos.

## Ejercicio 6 · Posición de un valor (búsqueda lineal)

Escribe `posicion(lista, objetivo)` que devuelva el **índice** de la primera
aparición de `objetivo`, o `-1` si no está.

**Ejemplo:** `posicion([10, 20, 30], 20)` → `1`; `posicion([10, 20], 99)` → `-1`.

Recorrer la lista dato por dato junto con su indice SI el dato actual ES IGUAL a objetivo ENTOCNES RETORNAR el indice actual  SI termina el recorrido y no encuentro el objetivo ENTONCES retornar -1

````python
def posicion(lista, objetivo):
    for indice in range (len(lista)):
        if lista[indice] == objetivo:
            return indice
    return -1
````

````python
revisar("encuentra en medio", posicion([10, 20, 30], 20) == 1)
revisar("primera posición", posicion([7, 7, 7], 7) == 0)
revisar("no está -> -1", posicion([10, 20], 99) == -1)
````

<details><summary>💡 Ver solución</summary>

```python

            return i           # salida temprana en la primera aparición

```
Recorremos por **índice** con `range(len(lista))` para poder devolver la posición.

## Ejercicio 7 · Días que subió la temperatura (comparar con el anterior)

Escribe `dias_que_subio(temps)` que cuente en cuántos días la temperatura fue
**mayor que la del día anterior**.

**Ejemplo:** `dias_que_subio([20, 22, 21, 25])` → `2` (subió 20→22 y 21→25).

PSEUDOCODIGO: Crear una variable contador y asignarle 0, recorrer la lista desde la segunda posicion hasta el final SI la temperatura en la posicion actual ES MAYOR a la temperatura de la posicion anterior ENTONCES  sumarle 1 al contadoor retornar contador

````python
def dias_que_subio(temps):
    contador = 0
    for i in range(1, len(temps)):
        if temps[i] > temps[i-1]:
            contador = contador +1
    return contador
````

````python
revisar("ejemplo", dias_que_subio([20, 22, 21, 25]) == 2)
revisar("siempre baja", dias_que_subio([5, 4, 3, 2]) == 0)
revisar("un solo día", dias_que_subio([10]) == 0)
````

<details><summary>💡 Ver solución</summary>

```python

    subidas = 0
    for i in range(1, len(temps)):   # desde el segundo día

            subidas += 1
```

## Ejercicio 8 · ¿Es palíndromo? (recorrer una cadena)

Escribe `es_palindromo(texto)` que devuelva `True` si `texto` se lee igual al
derecho y al revés. Considera el texto ya en minúsculas y sin espacios.

**Ejemplo:** `es_palindromo("reconocer")` → `True`; `es_palindromo("python")` → `False`.

Pista: en Python, `texto[::-1]` invierte una cadena.

PSEUDOCODIGO: crear una variable llamada texto_invertido SI texto ES IGUAL A texto invertido RETORNAR true SI NO RETORNAR false

````python
def es_palindromo(texto):
    # Palindromo es una palabra que se lee igual de izquierda a derecha y de derecha a izquierda
    texto_invertido = texto[::-1] #[::-1] estructura para que Python invierta un texto
    if texto == texto_invertido:
        return True
    else:
        return False
````

````python
revisar("reconocer", es_palindromo("reconocer") is True)
revisar("python", es_palindromo("python") is False)
revisar("una letra", es_palindromo("a") is True)
````

<details><summary>💡 Ver solución</summary>

```python
def es_palindromo(texto):
    return texto == texto[::-1]
```
`texto[::-1]` es el texto invertido; si coincide con el original, es palíndromo.

**Versión sin trucos** (recorriendo con dos índices), por si quieres ver la lógica:
