Implementa `clasificar(edad)` que devuelva:
- `"menor"` si `edad < 18`

- `"adulto mayor"` si `edad >= 65`

**Casos de ejemplo:** `clasificar(10)` → `"menor"`, `clasificar(40)` → `"adulto"`, `clasificar(70)` → `"adulto mayor"`.

```python
def clasificar(edad):
    if edad < 18:
        return "menor"
    elif edad  <= 65:
        return "adulto"
    else:
        return "adulto mayor"
    raise NotImplementedError("Implementa esta función")
```

```python
# === Tests adicionales (ocultos) · Ejercicio 1 ===
assert clasificar(17) == 'menor'
assert clasificar(18) == 'adulto'
assert clasificar(64) == 'adulto'
assert clasificar(65) == 'adulto mayor'
print("✅ Ejercicio 1: tests adicionales superados.")
```
