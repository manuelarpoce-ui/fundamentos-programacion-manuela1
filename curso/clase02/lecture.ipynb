## 1. De pseudocĂłdigo a Python: el rol de las variables

**ÂżQuĂŠ es una variable en Python?**

Una variable es un **nombre** que apunta a un **objeto en memoria**. Cuando
escribes `precio = 50000`, Python:
1. Crea un objeto de tipo `int` con el valor `50000` en algĂşn lugar de la memoria.
2. Hace que el nombre `precio` apunte a ese objeto.

```
   nombre          objeto en memoria
```

Esta distinciĂłn importa porque Python es un lenguaje de **tipado dinĂĄmico**:
el tipo lo lleva el objeto, no el nombre. La misma variable puede apuntar

### AsignaciĂłn mĂşltiple y convenciones de nombres


**Convenciones de nombres (PEP 8):**

## 2. Los cuatro tipos primitivos de Python

Python tiene cuatro tipos de datos *primitivos* o *atĂłmicos*: los que no se
descomponen en partes mĂĄs simples. Todo lo demĂĄs (listas, diccionarios,
objetos) son combinaciones de estos.

| Tipo | Nombre Python | Para quĂŠ sirve | Ejemplos |
|---|---|---|---|
| Entero | `int` | contar, indexar, aĂąos, cantidades | `0`, `42`, `-3`, `2024` |
| Decimal | `float` | precios, tasas, medidas | `3.14`, `1.5e6`, `-0.001` |
| Texto | `str` | nombres, categorĂ­as, mensajes | `"Bogota"`, `'tarjeta'` |
| Booleano | `bool` | flags, condiciones, resultados | `True`, `False` |

> đĄ En ciencia de datos, el tipo mĂĄs traicionero es `str`: los datos que
> llegan de CSV, bases de datos o APIs vienen *casi siempre* como texto,
> aunque representen nĂşmeros. Veremos cĂłmo convertirlos.

### 2b. Decimales (`float`)

Los `float` usan representaciĂłn **IEEE 754 de doble precisiĂłn** (64 bits).
Esto significa que pueden representar nĂşmeros con hasta ~15 dĂ­gitos
significativos, pero con limitaciones de precisiĂłn conocidas.

> â ď¸ Un resultado famoso: en Python `0.1 + 0.2` no es exactamente `0.3`. No es
> un bug: es una consecuencia de cĂłmo los decimales se guardan en binario. En
> aplicaciones financieras se usa el mĂłdulo `decimal` para dinero exacto.
> Para el curso, `float` es suficiente.

````python
# Decimales en contexto financiero.
tasa_iva    = 0.19          # 19% de IVA en Colombia
precio_base = 125000.0      # precio sin IVA
tasa_interes = 1.5e-2       # 1.5% mensual (notaciĂłn cientĂ­fica)

print("IVA (19%):", precio_base * tasa_iva)
print("Precio con IVA:", precio_base * (1 + tasa_iva))
print()

# La famosa sorpresa de los flotantes:
print("0.1 + 0.2 =", 0.1 + 0.2)                  # no es exactamente 0.3
print("0.1 + 0.2 == 0.3 es", 0.1 + 0.2 == 0.3)   # False (!)
print("Diferencia:", (0.1 + 0.2) - 0.3)            # ~5.5e-17 (ruido de precisiĂłn)
````

### 2d. Booleanos (`bool`)

`True` y `False` son los Ăşnicos valores booleanos. En Python, `bool` es
subclase de `int`: `True == 1` y `False == 0`, lo que da algunos trucos
Ăştiles (y algunas sorpresas).

Los booleanos son el resultado de comparaciones y expresiones lĂłgicas.
Son la **base de todas las decisiones** (`if`, `while`, `and`, `or`).

````python
# Booleanos: resultados de comparaciones.
monto         = 450000
es_mayorista  = monto > 200000     # True
tiene_descuento = False

print("ÂżEs venta mayorista?   ", es_mayorista,    type(es_mayorista))
print("ÂżTiene descuento?      ", tiene_descuento, type(tiene_descuento))
print()

# Truco: True y False son 1 y 0 (subclase de int).
aprobados = [True, True, False, True, False]
print("Total aprobados (suma de True):", sum(aprobados))   # 3

# OperaciĂłn con bool como int:
print("True + True =", True + True)     # 2 (!)
print("True * 5    =", True * 5)        # 5
````

## 3. ConversiĂłn de tipos (type casting)

En ciencia de datos, los datos casi siempre llegan *con el tipo incorrecto*:
los CSV traen todo como `str`, los formularios web tambiĂŠn. Saber convertir
tipos es una habilidad del dĂ­a a dĂ­a.

Python ofrece cuatro conversores explĂ­citos:

| FunciĂłn | Convierte a | Ejemplo | Resultado |
|---|---|---|---|
| `int(x)` | entero | `int("42")` | `42` |
| `float(x)` | decimal | `float("1.5")` | `1.5` |
| `str(x)` | texto | `str(99)` | `"99"` |
| `bool(x)` | booleano | `bool(0)` | `False` |

### ÂżCuĂĄndo convertir?

1. **Leer datos de CSV** â convertir texto a nĂşmero para operar.
2. **Mostrar resultados** â convertir nĂşmero a texto para concatenar.
3. **Comparar condiciones** â convertir a bool para usarlo en `if`.

````python
# Caso de uso real: datos de un formulario de venta.
# Todo llega como str (texto), necesitamos operar con ellos.

precio_str   = "85000"      # precio leĂ­do de un CSV
cantidad_str = "12"         # cantidad leĂ­da de un formulario
tasa_str     = "0.08"       # descuento leĂ­do como texto

# Convertimos antes de operar:
precio   = int(precio_str)
cantidad = int(cantidad_str)
tasa     = float(tasa_str)

subtotal  = precio * cantidad
descuento = subtotal * tasa
total     = subtotal - descuento

print("Subtotal:  $" + str(subtotal))
print("Descuento: $" + str(descuento))
print("Total:     $" + str(total))
````

### Reglas de `bool()`: quĂŠ es "falso" en Python

Python tiene una convenciĂłn potente: casi cualquier valor se puede evaluar
como booleano. Esta lista se llama **valores falsy**:

| Valor | bool() |
|---|---|
| `0`, `0.0` | `False` |
| `""` (cadena vacĂ­a) | `False` |
| `None` | `False` |
| `[]`, `{}`, `()` (colecciones vacĂ­as) | `False` |

**Todo lo demĂĄs** es `True` (truthy): cualquier nĂşmero distinto de cero,
cualquier cadena no vacĂ­a, cualquier colecciĂłn no vacĂ­a.

````python
# DemostraciĂłn de valores falsy:
print("bool(0)    =", bool(0))        # False
print("bool(0.0)  =", bool(0.0))      # False
print("bool('')   =", bool(''))       # False (cadena vacĂ­a)
print("bool(None) =", bool(None))     # False
print()
print("bool(1)    =", bool(1))        # True
print("bool(-99)  =", bool(-99))      # True (cualquier no-cero)
print("bool('0')  =", bool('0'))      # True (ÂĄcadena '0' NO es vacĂ­a!)
print("bool([])   =", bool([]))       # False (lista vacĂ­a)
print("bool([0])  =", bool([0]))      # True (lista con un elemento)
````

### Errores comunes de conversiĂłn

No toda conversiĂłn es posible: si el texto no se puede interpretar como
nĂşmero, Python lanza `ValueError`.

````python
# QuĂŠ pasa cuando la conversiĂłn falla:

# Esto funciona:
print(int("42"))
print(float("3.14"))

# Esto falla con ValueError:
try:
    int("cuarenta y dos")
except ValueError as e:
    print("Error con int('cuarenta y dos'):", e)

try:
    int("3.14")      # un float como str no se puede convertir directo a int
except ValueError as e:
    print("Error con int('3.14'):", e)

# La forma correcta para un float representado como str -> int:
valor = int(float("3.14"))
print("int(float('3.14')) =", valor)   # 3 (trunca)
````

### `None`: la ausencia de valor

`None` es el objeto especial de Python que representa **"ningĂşn valor"** o
**"aĂşn no definido"**. Es de tipo `NoneType` y es Ăşnico (hay un solo `None`
en memoria).

**CuĂĄndo aparece `None`:**
- Cuando una funciĂłn no tiene `return` explĂ­cito.
- Como valor por defecto para indicar "no aplica".
- Para inicializar variables antes de asignarles un valor real.

**CĂłmo comparar con `None`:**

```python
# Forma correcta: usar 'is' o 'is not'
if resultado is None:
    ...

# Evitar '==' con None (funciona pero es menos idiomĂĄtico):
if resultado == None:   # funciona, pero PEP 8 prefiere 'is'
    ...
```

````python
# None en el contexto de una funciĂłn que puede fallar.

def dividir_seguro(a, b):
    '''Devuelve a/b, o None si b es cero.'''
    if b == 0:
        return None    # seĂąal explĂ­cita de 'no se puede calcular'
    return a / b

resultados = [(10, 2), (5, 0), (100, 4)]
for a, b in resultados:
    r = dividir_seguro(a, b)
    if r is None:
        print("{}/{} = [indefinido: division por cero]".format(a, b))
    else:
        print("{}/{} = {:.2f}".format(a, b, r))
````

## 4. Operadores aritmĂŠticos

Python hereda la aritmĂŠtica estĂĄndar y aĂąade dos operadores muy usados en
programaciĂłn:

| Operador | Nombre | Ejemplo | Resultado |
|---|---|---|---|
| `+` | suma | `1000 + 200` | `1200` |
| `-` | resta | `5000 - 800` | `4200` |
| `*` | multiplicaciĂłn | `3 * 50000` | `150000` |
| `/` | divisiĂłn real | `7 / 2` | `3.5` (siempre float) |
| `//` | divisiĂłn entera | `7 // 2` | `3` (trunca) |
| `%` | mĂłdulo (residuo) | `7 % 2` | `1` |
| `**` | potencia | `2 ** 10` | `1024` |

Los dos mĂĄs novedosos son `//` (divisiĂłn entera) y `%` (mĂłdulo). Los viste
en el algoritmo del cajero en la Clase 1; hoy los entendemos a fondo.

````python
# Aritmetica bĂĄsica con contexto financiero.
precio_base = 350000
cantidad    = 3
tasa_iva    = 0.19
tasa_desc   = 0.05

subtotal       = precio_base * cantidad
descuento      = subtotal * tasa_desc
base_gravable  = subtotal - descuento
iva            = base_gravable * tasa_iva
total          = base_gravable + iva

print("Subtotal:      ${:,.0f}".format(subtotal))
print("Descuento (5%): ${:,.0f}".format(descuento))
print("Base gravable: ${:,.0f}".format(base_gravable))
print("IVA (19%):     ${:,.0f}".format(iva))
print("TOTAL:         ${:,.0f}".format(total))
````

### DivisiĂłn entera `//` y mĂłdulo `%`

Estos dos operadores son complementarios: juntos descomponen una divisiĂłn
en su cociente y residuo.

```
dividendo = divisor * cociente + residuo
   7       =    2   *    3     +    1

   Python:  7 // 2 = 3  (cociente)
            7 % 2  = 1  (residuo)
```

**Usos prĂĄcticos en finanzas y negocios:**
- `//` â convertir totales a unidades mayores (segundos a horas, pesos a billetes)
- `%` â saber el sobrante, detectar paridad, calcular ciclos

````python
# Uso 1: descomponer un tiempo en horas, minutos y segundos.
segundos_totales = 9753

horas   = segundos_totales // 3600
resto   = segundos_totales % 3600
minutos = resto // 60
segs    = resto % 60

print("{} segundos = {}h {}m {}s".format(segundos_totales, horas, minutos, segs))

print()

# Uso 2: repartir unidades en cajas (un problema de logĂ­stica).
unidades_pedido = 250
por_caja = 12

cajas_completas = unidades_pedido // por_caja
sobrante        = unidades_pedido % por_caja

print("Pedido de {} unidades en cajas de {}:".format(unidades_pedido, por_caja))
print("  {} cajas completas + {} unidades sueltas".format(cajas_completas, sobrante))

print()

# Uso 3: Âżes un nĂşmero par? (el mĂĄs clĂĄsico)
for n in [4, 7, 100, 33]:
    paridad = "par" if n % 2 == 0 else "impar"
    print("{}  ->  {}".format(n, paridad))
````

### Potencia `**` e interĂŠs compuesto

El operador `**` calcula potencias. En finanzas tiene una aplicaciĂłn directa:
el cĂĄlculo de **interĂŠs compuesto**.

```
Capital final = Capital inicial Ă (1 + tasa)^periodos
```

````python
# InterĂŠs compuesto: cuĂĄnto crece un capital con el tiempo.
capital_inicial = 1_000_000    # 1 millĂłn de pesos
tasa_mensual    = 0.015        # 1.5% mensual
periodos        = 12           # 12 meses

capital_final = capital_inicial * (1 + tasa_mensual) ** periodos
ganancia      = capital_final - capital_inicial

print("Capital inicial:  ${:,.0f}".format(capital_inicial))
print("Tasa mensual:      {:.1%}".format(tasa_mensual))
print("Periodos:          {} meses".format(periodos))
print("Capital final:    ${:,.0f}".format(capital_final))
print("Ganancia:         ${:,.0f}".format(ganancia))
print("Rendimiento total: {:.2%}".format(ganancia / capital_inicial))
````

### Trazado de ejecuciĂłn â verificando la aritmĂŠtica

Antes de confiar en un cĂĄlculo complejo, traza los valores intermedios.
Practiquemos con el cĂĄlculo de precio con IVA y descuento:

```
precio_base = 120000
descuento   = precio_base * 0.10  = 12000
```

### Cortocircuito: Python no evalĂşa mĂĄs de lo necesario

Con `and`, si la primera condiciĂłn es `False`, Python **no evalĂşa** la segunda
(porque el resultado ya es `False` pase lo que pase).

Con `or`, si la primera condiciĂłn es `True`, Python **no evalĂşa** la segunda.

Esto se llama **evaluaciĂłn perezosa** o **cortocircuito** (*short-circuit*).
Es importante cuando la segunda condiciĂłn es costosa o puede fallar.

```
Cortocircuito con and:
  False and (cualquier_cosa)  â  False  (no evalĂşa el segundo)

Cortocircuito con or:
  True  or  (cualquier_cosa)  â  True   (no evalĂşa el segundo)
```

````python
# DemostraciĂłn de cortocircuito.
# La funciĂłn division_peligrosa lanza un error si denominador es 0.
def division_peligrosa(a, b):
    return a / b   # ZeroDivisionError si b == 0

denominador = 0

# SIN cortocircuito (evaluarĂ­a ambos lados):
# if True and division_peligrosa(10, denominador):   # fallarĂ­a

# CON cortocircuito (and): si la izquierda es False, no se evalĂşa la derecha:
if denominador != 0 and division_peligrosa(10, denominador) > 1:
    print("Resultado positivo")
else:
    print("Denominador es cero; division_peligrosa NO fue llamada (cortocircuito).")

# CON cortocircuito (or): si la izquierda es True, no se evalĂşa la derecha:
ciudad_valida = True
if ciudad_valida or division_peligrosa(10, denominador) > 0:
    print("Condicion satisfecha sin evaluar el segundo operando.")
````

### Comparaciones encadenadas

## 6. Precedencia de operadores

Cuando una expresiĂłn mezcla varios operadores, Python sigue un orden de
evaluaciĂłn (como las matemĂĄticas con "primero multiplicaciĂłn que suma").

**Orden de mayor a menor precedencia (simplificado):**

```
1. ParĂŠntesis          ()
2. Potencia            **
3. Signo unario        -x, +x
4. MultiplicaciĂłn      *, /, //, %
5. Suma y resta        +, -
6. Comparaciones       ==, !=, <, >, <=, >=
7. not
8. and
9. or
```

**Regla prĂĄctica:** cuando tengas dudas, usa **parĂŠntesis**. Son gratuitos
en tĂŠrminos de rendimiento y hacen el cĂłdigo mĂĄs claro.

````python
# Precedencia en acciĂłn: mismos nĂşmeros, distintos resultados.
print("2 + 3 * 4    =", 2 + 3 * 4)     # 14 (no 20): * antes que +
print("(2 + 3) * 4  =", (2 + 3) * 4)   # 20: parĂŠntesis primero
print()

# Error clĂĄsico: calcular el promedio de dos precios.
precio_a = 80000
precio_b = 120000

# MAL: la divisiĂłn ocurre antes que la suma.
promedio_mal = precio_a + precio_b / 2
# BIEN: parĂŠntesis fuerzan la suma primero.
promedio_ok  = (precio_a + precio_b) / 2

print("Promedio MAL:", promedio_mal)   # 140000 (incorrecto)
print("Promedio OK: ", promedio_ok)    # 100000 (correcto)
print()

# Precedencia con lĂłgica:
monto = 180000
tiene_tarjeta = True

# ÂżSe aplica descuento? (>150000 Y tiene tarjeta) O (>300000)
aplica_1 = monto > 150000 and tiene_tarjeta or monto > 300000  # ambiguo
aplica_2 = (monto > 150000 and tiene_tarjeta) or monto > 300000  # explĂ­cito
print("Con parĂŠntesis explĂ­citos:", aplica_2)
````

### Reglas de oro para expresiones complejas

1. **Usa parĂŠntesis** cuando la precedencia no sea obvia.
2. **Divide expresiones largas** en varias variables intermedias.
3. **Traza** la expresiĂłn si no estĂĄs seguro del orden.

```python
# En lugar de esto:
total = precio * cantidad * (1 - descuento) * (1 + iva)

# Esto es mĂĄs trazable:
subtotal      = precio * cantidad
neto          = subtotal * (1 - descuento)
total         = neto * (1 + iva)
```

````python
# DemostraciĂłn: una sola lĂ­nea vs. variables intermedias.
precio    = 50000
cantidad  = 5
descuento = 0.10
iva       = 0.19

# Una lĂ­nea (difĂ­cil de depurar):
total_linea = precio * cantidad * (1 - descuento) * (1 + iva)

# Con variables intermedias (fĂĄcil de trazar):
subtotal = precio * cantidad
neto     = subtotal * (1 - descuento)
total_v  = neto * (1 + iva)

print("Total en una linea:   ${:,.2f}".format(total_linea))
print("Total con variables:  ${:,.2f}".format(total_v))
print("Son iguales:          ", abs(total_linea - total_v) < 0.01)
````

### Ejemplo completo: clasificaciĂłn de ventas

Apliquemos la metodologĂ­a de los 6 pasos.

**1. Contexto:** una tienda quiere etiquetar cada venta como "baja", "media"
o "alta" para reportes.

**2. SoluciĂłn intuitiva:** segĂşn el monto, elegimos la etiqueta.

**3. Algoritmo:**
1. Si monto > 200000 â etiqueta "alta"
2. Si no, pero monto > 50000 â etiqueta "media"
3. Si no â etiqueta "baja"

**4. PseudocĂłdigo:**
```
SI monto > 200000 ENTONCES
    categoria â "alta"
SINO SI monto > 50000 ENTONCES
    categoria â "media"
SINO
    categoria â "baja"
FIN_SI
```

````python
# 5. Python: traducciĂłn directa del pseudocĂłdigo.

def clasificar_venta(monto):
    # Clasifica una venta por su monto en COP.
    if monto > 200000:
        return "alta"
    elif monto > 50000:
        return "media"
    else:
        return "baja"

# Prueba con varios montos:
montos_prueba = [310000, 89000, 12000, 200000, 200001]
for m in montos_prueba:
    print("${:>8,}  ->  {}".format(m, clasificar_venta(m)))
````

### Ejemplo 2: descuento escalonado

En muchos comercios el descuento **aumenta con el monto**. Este patrĂłn
de decisiones escalonadas se llama **branching en cascada**.

**Reglas de negocio:**
- monto > 500000 â descuento del 15%
- monto > 200000 â descuento del 8%
- monto > 100000 â descuento del 3%
- monto <= 100000 â sin descuento

````python
def calcular_precio_final(precio_original):
    '''Aplica descuento escalonado segĂşn el monto.'''
    if precio_original > 500000:
        pct = 0.15
    elif precio_original > 200000:
        pct = 0.08
    elif precio_original > 100000:
        pct = 0.03
    else:
        pct = 0.00

    descuento = precio_original * pct
    precio_final = precio_original - descuento
    return precio_original, pct, descuento, precio_final

# Tabla de resultados:
print("{:<12} {:>8} {:>12} {:>12}".format("Original", "Descuento", "Valor desc.", "Final"))
print("-" * 50)
for monto in [80000, 150000, 250000, 600000]:
    orig, pct, desc, final = calcular_precio_final(monto)
    print("${:<10,} {:>7.0%} {:>12,.0f} {:>12,.0f}".format(orig, pct, desc, final))
````

### Ejemplo 3: aĂąo bisiesto

Un caso clĂĄsico de lĂłgica anidada con `and` y `or`.

**Regla:** un aĂąo es bisiesto si:
- Es divisible por 4, **Y**
- Si es divisible por 100, **tambiĂŠn** debe ser divisible por 400.

Esto cubre los casos especiales: 1900 (divisible por 100 pero no por 400 â
no bisiesto) y 2000 (divisible por 400 â bisiesto).

````python
def es_bisiesto(anio):
    '''Determina si un aĂąo es bisiesto segĂşn las reglas del calendario gregoriano.'''
    if anio % 400 == 0:
        return True
    elif anio % 100 == 0:
        return False
    elif anio % 4 == 0:
        return True
    else:
        return False

# Casos de prueba con explicaciĂłn:
casos = [(2000, True), (1900, False), (2024, True), (2023, False), (2100, False)]
for anio, esperado in casos:
    resultado = es_bisiesto(anio)
    marca = "OK" if resultado == esperado else "ERROR"
    print("{}: {}  [{}]".format(anio, resultado, marca))
````

### Ejemplo adicional: clasificar IMC

El **Ăndice de Masa Corporal** (IMC) es un indicador de salud usado en
medicina, seguros y app de bienestar.

```
IMC = peso_kg / altura_mÂ˛

< 18.5 â "bajo peso"
18.5 - 24.99 â "normal"
25 - 29.99 â "sobrepeso"
>= 30 â "obesidad"
```

````python
def clasificar_imc(peso_kg, altura_m):
    '''Calcula e interpreta el IMC.'''
    if altura_m <= 0:
        return None, "Error: altura debe ser positiva"
    imc = peso_kg / (altura_m ** 2)

    if imc < 18.5:
        categoria = "bajo peso"
    elif imc < 25:
        categoria = "normal"
    elif imc < 30:
        categoria = "sobrepeso"
    else:
        categoria = "obesidad"

    return round(imc, 1), categoria

# Prueba con varios perfiles:
perfiles = [(55, 1.70), (80, 1.75), (90, 1.65), (120, 1.72)]
print("{:<6} {:<8} {:>6} {:<12}".format("Peso", "Altura", "IMC", "Categoria"))
print("-" * 36)
for peso, altura in perfiles:
    imc, cat = clasificar_imc(peso, altura)
    print("{:<6} {:<8} {:>6} {:<12}".format(peso, altura, imc, cat))
````

### Ejemplo adicional: conversiĂłn de temperatura

Las conversiones entre escalas son ejercicios clĂĄsicos de operadores
aritmĂŠticos y tipos numĂŠricos.

```
Celsius a Fahrenheit:  F = C Ă (9/5) + 32
Celsius a Kelvin:      K = C + 273.15
```

````python
def convertir_temperatura(celsius):
    '''Convierte Celsius a Fahrenheit y Kelvin.'''
    fahrenheit = celsius * (9 / 5) + 32
    kelvin     = celsius + 273.15
    return fahrenheit, kelvin

# Tabla de conversiones:
print("{:>8} {:>12} {:>10}".format("Celsius", "Fahrenheit", "Kelvin"))
print("-" * 34)
for c in [-10, 0, 20, 37, 100]:
    f, k = convertir_temperatura(c)
    print("{:>8.1f} {:>12.1f} {:>10.2f}".format(c, f, k))
````

## 8. Operador ternario (expresiĂłn condicional)

Python tiene una forma compacta de escribir un `if/else` simple en una sola
lĂ­nea, cuando el resultado es una expresiĂłn (no un bloque de cĂłdigo):

```python
valor_si_true if condicion else valor_si_false
```

Este es el **operador ternario** o **expresiĂłn condicional**. Es equivalente
a:

```python
if condicion:
    resultado = valor_si_true
else:
    resultado = valor_si_false
```

> â Ăsalo para asignaciones simples. Si el bloque tiene mĂĄs de una lĂ­nea o
> la lĂłgica es compleja, prefiere el `if/else` tradicional: es mĂĄs legible.

````python
# Ternario: asignar una etiqueta segĂşn una condiciĂłn.
monto = 350000

# VersiĂłn if/else:
if monto > 200000:
    etiqueta = "mayorista"
else:
    etiqueta = "minorista"

# VersiĂłn ternaria (equivalente):
etiqueta2 = "mayorista" if monto > 200000 else "minorista"

print("Con if/else:   ", etiqueta)
print("Con ternario:  ", etiqueta2)
print("Son iguales:   ", etiqueta == etiqueta2)

# Otro ejemplo: calcular comisiĂłn del agente de ventas.
ventas = 5_200_000
comision = ventas * 0.05 if ventas >= 5_000_000 else ventas * 0.03
print()
print("Ventas: ${:,.0f}  -> Comision: ${:,.0f}".format(ventas, comision))
````

````python
# Ternario encadenado (con moderaciĂłn: puede volverse ilegible).
nota = 3.7

# Tres posibilidades: reprobado, aprobado, excelente.
estado = "excelente" if nota >= 4.5 else ("aprobado" if nota >= 3.0 else "reprobado")
print("Nota {}: {}".format(nota, estado))

# Comparar con if/elif/else para ver la diferencia de legibilidad:
if nota >= 4.5:
    estado2 = "excelente"
elif nota >= 3.0:
    estado2 = "aprobado"
else:
    estado2 = "reprobado"

print("Mismo resultado:", estado == estado2)
print("Para 3 ramas, el if/elif/else suele ser mas legible.")
````

### CĂłmo Python representa los tipos en memoria

Aunque no necesitamos gestionar la memoria manualmente, entender cĂłmo
Python organiza los objetos ayuda a evitar sorpresas.

```
  Tipo    | Ejemplo     | TamaĂąo en memoria (aprox.)
  --------|-------------|-----------------------------
  int     | 42          | 28 bytes (objeto Python base)
  float   | 3.14        | 24 bytes
  bool    | True        | 28 bytes (es subclase de int)
  str     | "hola"      | 49 + 1 byte por carĂĄcter
```

En Python, **todo es un objeto**. Incluso un simple `42` es un objeto con
su tipo, su contador de referencias y su valor. Esto hace que Python sea
flexible pero mĂĄs lento que C (donde `42` ocupa solo 4 bytes).

Para conjuntos grandes de datos numĂŠricos, **NumPy** usa arreglos de C
por dentro, ganando velocidad y eficiencia. Lo veremos en la Clase 5.

````python
import sys

# Cuanto 'pesa' cada tipo en Python:
print("sys.getsizeof(42):      ", sys.getsizeof(42), "bytes")
print("sys.getsizeof(3.14):   ", sys.getsizeof(3.14), "bytes")
print("sys.getsizeof(True):   ", sys.getsizeof(True), "bytes")
print("sys.getsizeof('hola'): ", sys.getsizeof('hola'), "bytes")
print("sys.getsizeof(''):     ", sys.getsizeof(''), "bytes")
print()
# Un int en Python pesa ~28 bytes; en C pestarĂ­a 4 u 8 bytes.
# Por eso, para millones de numeros, NumPy es hasta 10x mas eficiente.
print("ÂżPor que NumPy? Porque un entero Python es ~7x mas grande que en C.")
````

### Formatear cadenas: del `+` al `.format()` al f-string

Python tiene varias formas de combinar texto con variables. En el curso
usamos principalmente **f-strings** (Python 3.6+), que son los mĂĄs legibles:

```python
nombre = "Bogota"
ventas = 1350000

# ConcatenaciĂłn clĂĄsica (incĂłmoda, requiere str()):
"Ventas en " + nombre + ": $" + str(ventas)

# .format() (compatible con versiones antiguas):
"Ventas en {}: ${}".format(nombre, ventas)

# f-string (recomendado):
f"Ventas en {nombre}: ${ventas:,.0f}"
```

Los **especificadores de formato** controlan cĂłmo se muestra el nĂşmero:
- `{:,.0f}` â separador de miles, sin decimales
- `{:.2%}` â porcentaje con 2 decimales
- `{:>10}` â alineado a la derecha en 10 caracteres

````python
# Comparacion de estilos de formateo.
ciudad  = "Medellin"
ventas  = 2_350_000
tasa    = 0.192

# Los tres estilos producen el mismo resultado:
s1 = "Ventas en " + ciudad + ": $" + str(ventas)
s2 = "Ventas en {}: ${}".format(ciudad, ventas)
s3 = "Ventas en {}: ${:,.0f}".format(ciudad, ventas)

print(s1)
print(s2)
print(s3)
print()

# Tabla formateada con especificadores:
datos = [("Bogota", 3_200_000), ("Medellin", 2_350_000), ("Cali", 1_870_000)]
print("{:<12} {:>14}".format("Ciudad", "Ventas (COP)"))
print("-" * 28)
for c, v in datos:
    print("{:<12} {:>14,.0f}".format(c, v))
````

### Operadores de asignaciĂłn compuesta

Python tiene atajos para actualizar variables:

| Operador | Equivale a | Ejemplo |
|---|---|---|
| `x += n` | `x = x + n` | `total += precio` |
| `x -= n` | `x = x - n` | `stock -= cantidad` |
| `x *= n` | `x = x * n` | `capital *= (1 + tasa)` |
| `x /= n` | `x = x / n` | `precio /= 2` |
| `x //= n` | `x = x // n` | `horas //= 60` |
| `x %= n` | `x = x % n` | `resto %= divisor` |
| `x **= n` | `x = x ** n` | `base **= exponente` |

Son exactamente equivalentes a la forma larga; son solo una convenciĂłn.

````python
# Operadores compuestos en un ciclo de ventas del mes.
inventario  = 500
ventas_dia  = [12, 8, 15, 20, 5, 18, 9]
ingreso_total = 0
precio_unidad = 35000

for v in ventas_dia:
    inventario    -= v              # descontar del inventario
    ingreso_total += v * precio_unidad   # acumular ingreso

print("Ventas del mes:", sum(ventas_dia), "unidades")
print("Inventario restante:", inventario)
print("Ingreso total: ${:,.0f}".format(ingreso_total))
````
